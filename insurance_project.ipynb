```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv('/content/insurance.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df2=pd.get_dummies(df,drop_first=True)
```

```python
df2.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df2.drop('charges',axis=1),df2['charges'],test_size=.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
model=LinearRegression()
```

```python
train_model=model.fit(x_train,y_train)
```

```python
predict_charges=model.predict(x_test)
```

```python
predict_charges
```

```python
r2_score(y_test,predict_charges)
```

```python
mse=mean_squared_error(y_test,predict_charges)
```

```python
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
ridge=Ridge(alpha=.01)
```

```python
ridge.fit(x_train,y_train)
```

```python
ridge_predict=ridge.predict(x_test)
```

```python
ridge_predict
```

```python
r2_score(y_test,ridge_predict)
```

```python
mse=mean_squared_error(y_test,ridge_predict)
```

```python
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
lasso=Lasso(alpha=100)
```

```python
lasso.fit(x_train,y_train)
```

```python
lasso_predict=lasso.predict(x_test)
```

```python
lasso_predict
```

```python
r2_score(y_test,lasso_predict)
```

```python
mse=mean_squared_error(y_test,lasso_predict)
```

```python
mse
```

```python
rmse=np.sqrt(mse)
rmse
```
